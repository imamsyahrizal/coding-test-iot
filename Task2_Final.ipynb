```python
import pandas as pd
```

```python
dataSensorNew = pd.DataFrame(list(dataSensor['array']))
dataSensorNew
```

```python
dataSensorNew.info()
```

```python
df1=dataSensorNew.groupby(['roomArea', 'id'], as_index=False).sum().to_json(indent=4, orient='records')
#df1
```

```python
df1=pd.read_json(df1)
df1
```

```python
df2 = df1.drop('id', axis=1)
df2
```

### Min

```python
df_min=df2.groupby(['roomArea', pd.Grouper(freq='D', key='timestamp')]).min().reset_index()
df_min=df_min.rename(columns={'temperature':'temperatureMin', 'humidity':'humidityMin'})
dfTemp_min = df_min.drop('humidityMin', axis=1)
dfHum_min = df_min.drop('temperatureMin', axis=1)
```

### Max

```python
df_max = df2.groupby(['roomArea', pd.Grouper(freq='D', key='timestamp')]).max().reset_index()
df_max=df_max.rename(columns={'temperature':'temperatureMax', 'humidity':'humidityMax'})
dfTemp_max = df_max.drop('humidityMax', axis=1)
dfHum_max = df_max.drop('temperatureMax', axis=1)
```

### Median

```python
df_median=df2.groupby(['roomArea', pd.Grouper(freq='D', key='timestamp')]).median().reset_index()
df_median=df_median.rename(columns={'temperature':'temperatureMedian', 'humidity':'humidityMedian'})
dfTemp_median = df_median.drop('humidityMedian', axis=1)
dfHum_median = df_median.drop('temperatureMedian', axis=1)
```

### Average

```python
df_mean=df2.groupby(['roomArea', pd.Grouper(freq='D', key='timestamp')]).mean().reset_index()
df_mean=df_mean.rename(columns={'temperature':'temperatureMean', 'humidity':'humidityMean'})
dfTemp_mean = df_mean.drop('humidityMean', axis=1)
dfHum_mean = df_mean.drop('temperatureMean', axis=1)
```

## All

```python
dfTemperature= pd.merge(dfTemp_min, dfTemp_max, on=['roomArea', 'timestamp']).merge(dfTemp_median, on=['roomArea', 'timestamp']).merge(dfTemp_mean, on=['roomArea', 'timestamp'])
dfTemperature=dfTemperature.rename(columns={'temperatureMin':'Min', 'temperatureMax':'Max', 'temperatureMedian':'Median', 'temperatureMean':'Mean'})
dfTemperature
```

```python
dfHumidity= pd.merge(dfHum_min, dfHum_max, on=['roomArea', 'timestamp']).merge(dfHum_median, on=['roomArea', 'timestamp']).merge(dfHum_mean, on=['roomArea', 'timestamp'])
dfHumidity=dfHumidity.rename(columns={'humidityMin':'Min', 'humidityMax':'Max', 'humidityMedian':'Median', 'humidityMean':'Mean'})
dfHumidity
```
